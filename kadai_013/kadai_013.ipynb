```python
# 消費税を加えた計算結果を返す関数を記述
def calculate_total(price: int, tax_rate: int) -> float:
    # 消費税額を計算（商品の金額＋商品の金額×消費税÷100））
    total_price = price + price * tax_rate / 100

    # 計算結果を戻り値として返す
    return total_price

# 第1引数に商品の金額「110」、第2引数に消費税（10%）の「10」を設定して関数を呼び出す
result = calculate_total(110, 10)

# 計算結果を出力する
print(f"{result}円")
```
